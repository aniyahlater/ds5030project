```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("hf://datasets/ailsntua/Chordonomicon/chordonomicon_v2.csv")
```

```python
songdb = sqlite3.connect('songs.db')
```

```python
df.to_sql('songs', songdb, index=False, chunksize=1000, if_exists='replace')
```

```python
rnb_query = '''
SELECT *
FROM songs
WHERE genres LIKE '%%r&b%%';
'''

rnb = pd.read_sql_query(rnb_query, con=songdb)
rnb.head()
```
